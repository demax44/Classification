Make a list and iterate

```
years=list(range(1880,1971))
years
```

```
Baby_namesyears=[]
for year in years:
  BabyData=pd.read_csv('/content/yob{}.txt'.format(year),header=None,names=["Name","Gender","Count"])
  Baby_namesyears.append(BabyData)
```

```
Baby_namesyears
```

```
len(Baby_namesyears)
```

```
BabyB=pd.concat(Baby_namesyears,axis=0,keys=years,names=['Year']).droplevel(-1).reset_index()
```

```
BabyB
```

```
BabyB.to_csv('US_infantnames.csv',index=False)
```
