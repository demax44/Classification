```
def result_finance(rewards):
  result_finance=rewards.agg(['mean','std']).T
  result_finance["Return"]=result_finance['mean']*252
  result_finance['Risk']=result_finance['std']*np.sqrt(252)
  result_finance.drop(columns=['mean','std'],inplace=True)
  return result_finance
```
